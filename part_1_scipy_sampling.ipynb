This notebook depicts ways to sample a probaility distribution using [SciPy's stats](https://docs.scipy.org/doc/scipy/reference/stats.html) utilities. 

```python
import numpy as np, scipy
from matplotlib import pyplot
from open_atmos_jupyter_utils import show_plot

Ï_geom = 1.4
median_size = 1
n_sd = 64

pyplot.hist(scipy.stats.lognorm.ppf(
    np.linspace(.001, .999, n_sd),
    np.log(Ï_geom),
    0,
    median_size,
), label="constant-multiplicity sampling", density=True)

sizes = np.linspace(0, 3, n_sd)
pyplot.plot(sizes, scipy.stats.lognorm.pdf(
    sizes,
    np.log(Ï_geom),
    0,
    median_size,
), label="uniform pdf sampling", marker='o')

pyplot.legend(loc='upper right')
pyplot.gca().set(ylabel="super droplet count", xlabel="size parameter")
show_plot()
```
